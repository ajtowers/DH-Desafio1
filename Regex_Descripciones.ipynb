```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('properatti_clean.csv')
df.head(2)
```

```python
#Cantidad de Observaciones por Variable
df.info()
```

Diccionario para mapear los textos a numeros

```python
numeros_en_letras = {'uno' : '1' , 'dos' : '2' , 'tres' : '3' , 'cuatro' : '4' , 'cinco' : '5' , 'seis' : '6' , 'siete' : '7' , 'mono' : '1' , 'mon' : '1' , 'nan' : np.nan}
```

Aca seteamos la REGEX para extraer la cantidad de ambientes

```python
pattern = r'(uno\s|dos\s|tres\s|cuatro\s|cinco\s|seis\s|siete\s|\d+?\s|mono?)amb{0,1}'
df['ambientes'] = ""
df['ambientes'] = df['descripcion'].str.extract(pattern , flags=re.IGNORECASE , expand=True)
```

```python
check = df
print("Cantidad de Observaciones nuevas --> " , check[(check['habs'].isnull()) & (check['ambientes'].notnull())].count()['ambientes'])
```

```python
df['ambientes'] = df['ambientes'].str.lower()
df['ambientes'].unique()
```

```python
df['ambientes'] = df['ambientes'].astype('str')
df['ambientes'].dtypes
```

```python
df['ambientes'] = df['ambientes'].str.strip()
df['ambientes'].unique()
```

Reemplazamos los numeros en 'letras' a numeros en digitos

```python
df['ambientes'].replace(numeros_en_letras , inplace=True)
df['ambientes'].unique()
```

```python
df.head(100)
```

```python
df['habitaciones_final'] = df['habs']
df['habitaciones_final'] = df['habitaciones_final'].fillna(df['ambientes'])
df.info()
```

```python
df['habitaciones_final'] = df['habitaciones_final'].dropna().astype('int64')
```

Regex para info de los deptos

```python
amenidades = 'r(balcon)|(cochera)|(patio)|(gimnasio|gym)|(terraza)|(sum)|(baulera)|(parrilla)'
df_amen = df['descripcion'].str.extract(amenidades , flags=re.IGNORECASE , expand=True)
df_amen.columns = ['balcon' , 'cochera' , 'patio' , 'gym' , 'terraza' , 'sum' , 'baulera' , 'parrilla']

df_amen.fillna(0 , inplace=True)
df_amen.head()

df_amen = df_amen.applymap(lambda x: 1 if len(str(x)) > 1 else 0)
df_amen.head()
```

```python
df_amen.head(100)
```

```python
df = df.join(df_amen)
```
