```python
props_df = pd.read_csv('properatti_clean.csv')
```

```python
props_df.head()
```

```python
words = {}
# table = {ord(char): None for char in ',.-+:'}
for des in props_df.descripcion:
    texto = des.split()
    for word in texto:
        word = word.lower()      #.translate(table)
        word = re.sub(r'[^a-z0-9]', r'', word)
        if word in words:
            words[word] += 1
        else:
            words[word] = 1
```

```python
list_words = [(k,v) for k,v in words.items()]
sorted_words = sorted(list_words, key=lambda x: x[1], reverse=True)
#sorted_words = [x for x,y in sorted_words if y>2400]
sorted_words = [x for x,y in sorted_words]
sorted_words
```

```python
amenities = ['pileta', 'seguridad', 'parrilla', 'balcon', 'cochera', 'nuevo']
```

```python
# funcion que recibe un objeto string, y matchea
# palabras vs regex en un df predefinido

def split_clean_build(x, keys):
    
    # inicializo el dataframe que va a tener los datos de las cols
    d = {col: 0 for col in keys}
    
    frase = x.lower()
    
    # matches
    if re.search(r'pile|pisc', frase): 
        d[keys[0]] = 1
    if re.search(r'segur|vigil', frase): 
        d[keys[1]] = 1
    if re.search(r'parri', frase): 
        d[keys[2]] = 1
    if re.search(r'balcon|balcón', frase): 
        d[keys[3]] = 1
    if re.search(r'coch', frase): 
        d[keys[4]] = 1
    if re.search(r'pozo|nuev', frase): 
        d[keys[5]] = 1
    
    return pd.DataFrame.from_dict(d, orient='index').T
    
```

```python
dummy = pd.DataFrame(columns=amenities)

for row in props_df.descripcion:
    line = split_clean_build(row, amenities)
    dummy = pd.concat([dummy, line])
    
dummy.reset_index(inplace=True)
```

```python
props_df = props_df.join(dummy)
```

```python
props_df.drop(['index'], axis=1, inplace=True)
```
