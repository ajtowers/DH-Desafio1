```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Lectura del dataset usando como index la primera columna
df = pd.read_csv('properatti_clean.csv', index_col=0)
```

```python
df.head()
df.shape
```

```python
descriptions = df.description
descriptions.head()
```

```python
words = {}
for des in descriptions:
    texto = des.split()
    for word in texto:
        word = word.lower()
        if word in words:
            words[word] += 1
        else:
            words[word] = 1
```

```python
list_words = [(k,v) for k,v in words.items()]
sorted_words = sorted(list_words, key=lambda x: x[1], reverse=True)
sorted_words = [ (x, y) for x,y in sorted_words if y>2400]
sorted_words[15:]
```
